```python
stopwords = ['我', '對','與',' ','的','在','是','有','了','說','她','也','就','不','和','臺','長','為','去','之','後','於','但','人','一個','他','來','最','都','而','到']
puncs = '！？｡＂＃＄％＆＇（）＊＋，－／：；＜＝＞＠［＼］＾＿｀｛｜｝～｟｠｢｣､、〃》「」『』【】〔〕〖〗〘〙〚〛〜〝〞〟〰〾〿–—‘’‛“”„‟…‧﹏.。'
```

```python
seg_list = jieba.cut(text, cut_all=False)  # 精确模式
seg_list = list(seg_list)
wordDict = {}
for w in seg_list:
    if (w not in stopwords) and (w not in puncs):
        if w in wordDict:
            wordDict[w] = wordDict[w] + 1
        else:
            wordDict[w] = 1
print(wordDict)
```

```python
result = pd.DataFrame({'key':[x for x in wordDict.keys()],'value':[x for x in wordDict.values()]})
result.head(10)
```

```python
result = result.sort_values(by='value',ascending=False)
result = result.reset_index(drop=True)
result.head(20)
```

```python
result.to_csv("CMKuan_wordcloud.csv", encoding='utf-8')
```
